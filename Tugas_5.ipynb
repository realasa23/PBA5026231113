```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.metrics.pairwise import cosine_similarity
import pandas as pd

# 1. Dataset 3 Kalimat Artikel Raja Ampat
corpus = [
    "Temuan Greenpeace, aktivitas tambang di Pulau Gag merusak ekosistem hutan dan laut di sekitarnya.",
    "Aktivitas pertambangan nikel di Kepulauan Raja Ampat, Papua Barat Daya menimbulkan kerusakan ekologi, mulai dari kerusakan pesisir sampai terumbu karang hingga deforestasi yang memicu krisis iklim.",
    "Banyak ekosistem laut rusak akibat tambang, dari hasil tangkapan ikan yang berkurang, budidaya rumput laut yang gagal, hingga terumbu karang yang hancur."
]

# 2. Representasi Vektor (Bag of Words)
# Fokus pada kata kunci utama yang muncul di beberapa kalimat agar tabel rapi seperti contoh
vocab_pilihan = ['tambang', 'ekosistem', 'laut', 'terumbu', 'karang']
vectorizer = CountVectorizer(vocabulary=vocab_pilihan)
X = vectorizer.fit_transform(corpus)

# 3. Hitung Cosine Similarity
# Membandingkan semua dokumen dengan Dokumen 0 (Kalimat 1)
cos_sim = cosine_similarity(X[0:1], X)

# 4. Tampilkan Hasil ke dalam DataFrame Pandas
df = pd.DataFrame(X.toarray(), columns=vocab_pilihan)

# Menambahkan nama baris agar mudah dibaca
df.insert(0, 'Dokumen', ['Kalimat 1', 'Kalimat 2', 'Kalimat 3'])
df['Cosine Sim to Kalimat 1'] = cos_sim.flatten()

# Print hasilnya
print(df.to_string(index=False))
```

**CountVectorizer:** Mengubah teks menjadi angka (matriks frekuensi kata). Di sini kita membatasi fitur hanya pada kata "hate" dan "darkness" agar sesuai dengan grafik di slide Stanford.

**Cosine Similarity vs Distance:** Fungsi ini menghasilkan nilai 0 sampai 1. Semakin mendekati 1, semakin mirip kedua dokumen tersebut.

**Analisis Hasil:** Bahwa Dokumen 2 (MLK) memiliki nilai similarity yang tinggi terhadap Dokumen 0, membuktikan bahwa meskipun jumlah katanya jauh lebih banyak, "arah" topiknya tetap serupa.

**Euclidean Distance:** Jika kita menggunakan jarak lurus, Dokumen 2 akan dianggap sangat berbeda karena jumlah kemunculan katanya (3 dan 4) jauh dari Dokumen 0 (1 dan 2).

**Aplikasi di Dunia Nyata:** Teknik ini adalah dasar dari mesin pencari (search engines) dan sistem rekomendasi teks yang sering kita gunakan sehari-hari.

```python
import pandas as pd
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.metrics.pairwise import cosine_similarity

# 1. Dataset 3 Kalimat Artikel Raja Ampat
documents = [
    "Temuan Greenpeace, aktivitas tambang di Pulau Gag merusak ekosistem hutan dan laut di sekitarnya.",
    "Aktivitas pertambangan nikel di Kepulauan Raja Ampat, Papua Barat Daya menimbulkan kerusakan ekologi, mulai dari kerusakan pesisir sampai terumbu karang hingga deforestasi yang memicu krisis iklim.",
    "Banyak ekosistem laut rusak akibat tambang, dari hasil tangkapan ikan yang berkurang, budidaya rumput laut yang gagal, hingga terumbu karang yang hancur."
]

# 2. Membuat Matriks Term Frequency (TF) untuk SEMUA kata
vectorizer = CountVectorizer()
X = vectorizer.fit_transform(documents)

# Menyimpan hasil TF ke dalam DataFrame Pandas agar mudah dibaca
tf = pd.DataFrame(X.toarray(), columns=vectorizer.get_feature_names_out())
tf.index = ['Kalimat 1', 'Kalimat 2', 'Kalimat 3']

print("--- Matriks Term Frequency (TF) ---")
# Menampilkan 10 kolom kata pertama saja agar tampilan terminal tidak terlalu panjang
print(tf.iloc[:, :10].to_string())
print("... (dan seterusnya untuk semua kata)\n")

# 3. Menghitung Full Cosine Similarity Matrix
full_cos_sim_matrix = cosine_similarity(X)

# Menampilkan hasil perhitungan Cosine Similarity ke dalam DataFrame
cos_sim_df = pd.DataFrame(
    full_cos_sim_matrix,
    index=['Kalimat 1', 'Kalimat 2', 'Kalimat 3'],
    columns=['Kalimat 1', 'Kalimat 2', 'Kalimat 3']
)

print("--- Full Cosine Similarity Matrix ---")
print(cos_sim_df)
```
