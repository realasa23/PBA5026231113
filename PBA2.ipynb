```python
import nltk
from nltk.tokenize import word_tokenize
from collections import Counter

nltk.download("gutenberg")
nltk.download("punkt")
nltk.download("punkt_tab") # Add this line to download the missing resource

# 1) Ambil salah satu buku dari Gutenberg corpus
file_id = "milton-paradise.txt"
text = nltk.corpus.gutenberg.raw(file_id)

# 2) (Opsional) ambil potongan supaya cepat
text_snippet = text[:2000]  # ambil 2000 karakter pertama

# 3) Tokenisasi
tokens = word_tokenize(text_snippet)

print("File:", file_id)
print("20 token pertama:", tokens[:20])
print("Total token:", len(tokens))
print("Token unik:", len(set(tokens)))

# 4) Simpan token ke file
with open("gutenberg_tokens.txt", "w", encoding="utf-8") as f:
    f.write("\n".join(tokens))
print("Disimpan: gutenberg_tokens.txt")

# 5) Frekuensi token (top 10)
freq = Counter([t.lower() for t in tokens]).most_common(10)
print("Top 10 token:", freq)
```

```python
import pandas as pd

columns_name = ['text', 'label']
df_train = pd.read_csv('../data/raw/train_preprocess.tsv', sep='\t', header=None, names=columns_name)
df_train.head()
```

```python
# Pilih beberapa contoh teks
sample_texts = df_train['text'][:3].tolist()  # Ambil 3 teks pertama
labels = df_train['label'][:3].tolist()  # Ambil label sentimen
```

```python
import os

stemmer = StemmerFactory().create_stemmer()

# Pastikan direktori output ada
os.makedirs('../data/processed', exist_ok=True)

# Tokenisasi dan analisis untuk setiap teks
for i, text in enumerate(sample_texts):
    print(f"\nContoh Teks {i+1} (Label: {labels[i]}):")
    print("Teks Asli:", text)

    # Tokenisasi
    tokens = word_tokenize(text)
    print("Token:", tokens)
    print("Jumlah Token:", len(tokens))
    print("Token Unik:", len(set(tokens)))

    # Stemming (opsional untuk analisis)
    stemmed_text = stemmer.stem(text)
    stemmed_tokens = word_tokenize(stemmed_text)
    print("Token setelah Stemming:", stemmed_tokens)
    print("Jumlah Token setelah Stemming:", len(stemmed_tokens))

    # Simpan token ke file
    with open(f'../data/processed/tokens_sample_indonlu_smsa_{i+1}.txt', 'w', encoding='utf-8') as f:
        f.write('\n'.join(tokens))
    print(f"Token disimpan ke tokens_sample_indonlu_smsa_{i+1}.txt")
```

```python
!pip -q install nltk pandas

import nltk
from nltk.tokenize import word_tokenize
from nltk.stem import PorterStemmer
import pandas as pd
from collections import Counter

nltk.download("gutenberg")
nltk.download("punkt")
nltk.download("punkt_tab") # Add this line to download the missing resource

# 1) Load text dari Gutenberg
file_id = "austen-emma.txt"
text = nltk.corpus.gutenberg.raw(file_id)

# 2) Bikin "dataset" sederhana: ambil 3 potongan teks (misal per 800 karakter)
chunk_size = 800
sample_texts = [text[i:i+chunk_size] for i in range(0, chunk_size*3, chunk_size)]
labels = [file_id] * 3  # label opsional (misal nama file)

# 3) Stemmer Inggris (Porter)
stemmer = PorterStemmer()

# 4) Tokenisasi + analisis
for i, t in enumerate(sample_texts):
    print(f"\nContoh Teks {i+1} (Label: {labels[i]}):")
    print("Teks Asli:", t[:200], "...")  # biar ga kepanjangan

    tokens = word_tokenize(t)
    print("Token:", tokens[:30], "...")
    print("Jumlah Token:", len(tokens))
    print("Token Unik:", len(set(tokens)))

    # Stemming (Porter)
    stemmed_tokens = [stemmer.stem(tok) for tok in tokens]
    print("Token setelah Stemming:", stemmed_tokens[:30], "...")
    print("Jumlah Token setelah Stemming:", len(stemmed_tokens))

    # Simpan token ke file
    with open(f"tokens_sample_{i+1}.txt", "w", encoding="utf-8") as f:
        f.write("\n".join(tokens))
    print(f"Token disimpan ke tokens_sample_{i+1}.txt")

# 5) Frekuensi kata top 10
all_tokens = []
for t in sample_texts:
    all_tokens.extend([tok.lower() for tok in word_tokenize(t)])

word_freq = Counter(all_tokens).most_common(10)
print("\n10 Kata Teratas:", word_freq)
```
