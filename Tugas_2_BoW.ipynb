#BoW bag of Words


# Step 1: Tokenize the data, remove stop words and perform stemming or lemmatization.

```python
import pandas as pd
import nltk
import re
from collections import Counter
from nltk.stem import WordNetLemmatizer
from nltk.corpus import stopwords
from sklearn.feature_extraction.text import CountVectorizer

# Download required NLTK datasets
nltk.download('stopwords')
nltk.download('punkt')
nltk.download('wordnet')

url = "https://raw.githubusercontent.com/realasa23/PBA5026231113/c0208eba279ec6e30446f6254ffeee0ea70480bb/review_flip.xlsx"

# Membaca 50 baris pertama
df = pd.read_excel(url, nrows=50)

# Ganti 'content' dengan nama kolom yang berisi teks review di file Excel-mu.
# (Misalnya: 'review', 'text', 'ulasan', dll)
nama_kolom = 'content'

# Mengambil kolom teks, menghapus baris yang kosong, dan mengubahnya menjadi list
reviews = df[nama_kolom].dropna().astype(str).tolist()


lemmatizer = WordNetLemmatizer()
stop_words = set(stopwords.words('english'))
corpus = []

for review in reviews:
    # Menghapus karakter selain huruf
    sent = re.sub('[^a-zA-Z]', ' ', review)
    sent = sent.lower()
    sent = sent.split()

    # Lemmatization & Stopword Removal
    sent = [lemmatizer.lemmatize(word) for word in sent if word not in stop_words]
    sent = ' '.join(sent)
    corpus.append(sent)

print("--- HASIL STEP 1 (Corpus) ---")
print(corpus[:3], "...\n") # Print 3 baris pertama saja
```

# STEP 2: List All Unique Words

```python
all_words = []
for sentence in corpus:
    # FIX: Tambahkan loop untuk memecah kalimat menjadi kata
    for word in sentence.split():
        all_words.append(word)

unique_words = list(set(all_words))
print("--- HASIL STEP 2 (Unique Words) ---")
print("Total kata unik:", len(unique_words))
print("Contoh kata unik:", unique_words[:10], "...\n")
```

# STEP 3: Mapping Words to Numbers & Frequencies

```python
word_frequencies = Counter(all_words)

# Sorting words by frequency (descending)
sorted_words = sorted(word_frequencies.items(), key=lambda item: item[1], reverse=True)

# Creating the dictionary mapping words to numbers
word_to_number = {}
for i, (word, frequency) in enumerate(sorted_words):
    word_to_number[word] = i + 1  # Assign numbers starting from 1

print("--- HASIL STEP 3 (Frequencies & Mapping) ---")
print("Top 5 Word Frequencies:", sorted_words[:5])
print("Contoh Mapping Word to Number:", {k: word_to_number[k] for k in list(word_to_number)[:5]}, "...\n")
```


# STEP 4: CountVectorizer (Bag of Words)


```python
cv = CountVectorizer()
independentFeatures = cv.fit_transform(corpus).toarray()

print("--- HASIL STEP 4 (Bag of Words Array) ---")
print("Dimensi Array BoW:", independentFeatures.shape)
print("Isi Array (3 baris pertama):\n", independentFeatures[:3])
```
